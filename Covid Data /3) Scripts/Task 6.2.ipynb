```python
# Importing cleaned dataset
Covid_data = pd.read_csv(os.path.join(path, 'Data', 'Prepared data', 'Covid_data.csv'))
```

```python
# Dropping unneeded columns and creating data subset
Covid_Data = Covid_data.drop(['Unnamed: 0','Country','Continent'], axis = 1)
```

```python
Covid_Data.head()
```

### Creating heatmap

```python
# Creating correlation table between variables
Covid_Data.corr()
```

```python
# Creating subplot using matplotlib
f,ax = plt.subplots(figsize=(10,10))

# Creating matrix heatmap using seaborn
corr = sns.heatmap(Covid_Data.corr(), annot = True, ax = ax)
```

```python
#exporting the visual
corr.figure.savefig(os.path.join(path, '4) Analysis','Visualizations', 'Covid_heatmap.png'))
```

Population and Total covid-19 cases are somewhat positivily related with corrleation of 0.55. Also, Total Covid cases are highly positivily related to critical case, active cases and total dealth. So we can conclude that high population countries has higher chances of having Covid case which will automatically have more active case, critical case and total dealth    

## Creating Scatterplot

```python
# Scatterplot of most correlated variables
# Population cases and Total Cases
sns.lmplot(x = 'Population', y = 'Total Cases', data = Covid_Data)
```

```python
# Total Cases and Critical Cases
sns.lmplot(x = 'Total Cases', y = 'Critical Cases', data = Covid_Data)
```

```python
# Total Cases and Total Deaths
sns.lmplot(x = 'Total Cases', y = 'Total Deaths', data = Covid_Data)
```

```python
# Total Cases and Total Tests
sns.lmplot(x = 'Total Cases', y = 'Total Tests', data = Covid_Data)
```

The scatterplots illustrate the strong correlations that we see between Population, total cases, total death, Critical cases and Total tests all overall the countries.

## Creating a Pair Plot

```python
# Creating a pairplot
pair_plot = sns.pairplot(Covid_Data)
```

```python
#exporting the visual
pair_plot.figure.savefig(os.path.join(path, '4) Analysis','Visualizations', 'Covid_correlation1.png'))
```

## Creating a Categorical Plot

```python
# Creating a histogram to see distribution
sns.histplot(Covid_Data['Total Cases'], bins = 10, kde = True)
```

```python
# Creating category for categorical plot
Covid_Data.loc[Covid_Data['Deaths/1M pop'] < 88 , 'Category'] = 'Less Vulnerable'
```

```python
# Creating category for categorical plot
Covid_Data.loc[Covid_Data['Deaths/1M pop'] > 88 , 'Category'] = 'Vulnerable'
```

```python
Covid_Data['Category'].value_counts(dropna = False)
```

```python
# Creating categorical plot with Population
sns.set(style="ticks")
cat_plot_Population = sns.catplot(x='Population', y='Deaths/1M pop', hue="Category", data=Covid_Data)
```

```python
#exporting the visual
cat_plot_Population.figure.savefig(os.path.join(path, '4) Analysis','Visualizations', 'population_cluster.png'))
```

```python
# Creating categorical plot with Total Cases
sns.set(style="ticks")
cat_plot_Total_Cases = sns.catplot(x='Total Cases', y='Deaths/1M pop', hue="Category", data=Covid_Data)
```

```python
#exporting the visual
cat_plot_Total_Cases.figure.savefig(os.path.join(path, '4) Analysis','Visualizations', 'TotalCases_cluster.png'))
```
